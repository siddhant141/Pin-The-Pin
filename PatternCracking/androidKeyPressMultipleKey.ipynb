```
import pandas as pd
import matplotlib.pyplot as plt
import csv
import os
```

```
def make_graph(filepath , sensor):
  t = []
  x = []
  y = []
  z = []
  with open(filepath ,'r') as csvfile:
      plots = csv.reader(csvfile, delimiter=',')
      for row in plots:
          if(row[0] != 'Time'):
            t.append(float(row[0]))
            x.append(float(row[1]))
            y.append(float(row[2]))
            z.append(float(row[3]))
  plt.plot(t,x, label='X')
  plt.plot(t,y, label='Y')
  plt.plot(t,z, label='Z')

  plt.xlabel('Time')
  plt.ylabel('X')
  plt.title( sensor+' Graph\nFor finding pressed key')
  plt.legend()

  plt.show()
```

```

filepath = '/content/drive/My Drive/AndroidProject/AccApp'

files = []
# r=root, d=directories, f = files
for r, d, f in os.walk(filepath):
    for file in f:
        if '.csv' in file:
            files.append(os.path.join(r, file))

for f in files:
    print(f)
    df = pd.read_csv(f)
    print(df.head())
    count = 0
    words = f.split("/")
    sensor = words[6]
    print(sensor)
    make_graph(f , sensor)
```
