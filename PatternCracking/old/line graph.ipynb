```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import csv
from sklearn.cluster import KMeans
```

```python
path='C:\\Users\\nilu1\\PatternCracking\\gyro\\11-Dec-2019_120141_PM.csv'
f=open(path)
para=f.readlines()
line=""
data=[]
for i in range(len(para)):
    line=para[i]
    line=line.split(",")
    for i in range(len(line)):
        line[i]=line[i].rstrip('\n')
        line[i]=line[i].strip('"')
        line[i]=float(line[i])
    data.append(line)
X=np.array(data)
```

```python
plt.plot(X[:,0],X[:,1])
plt.ylabel('x - axis') 
plt.xlabel('time')  
plt.title('X-axis') 
plt.show() 
```

```python
plt.plot(X[:,0],X[:,2])
plt.ylabel('y - axis') 
plt.xlabel('time')  
plt.title('Y-axis') 
plt.show() 
```
