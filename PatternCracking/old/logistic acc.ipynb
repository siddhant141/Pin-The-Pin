```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn import metrics
from os import listdir
from os.path import isfile, join
```

```python
path="C:\\Users\\nilu1\\PatternCracking\\acc\\"
pathex=[]
temp = [f for f in listdir(path) if isfile(join(path, f))]
temp.sort()
for i in range(len(temp)):
    st=temp[i]
    if(st[:2]=="27"):
        pathex.append(temp[i])
# print(pathex)
```

```python
# n -> number of lines in csv file
# actual -> the actual key pressed
n=100
actual=1
x=[]
y=[]
z=[]
temp=pd.read_csv(path+pathex[0])
temp=temp.head(n)
x=list(temp["X"])
y=list(temp["Y"])
z=list(temp["Z"])
maxx=max(x)
minx=min(x)
maxy=max(y)
miny=min(y)
maxz=max(z)
minz=min(z)
data=[[maxx-x[0],minx-x[0],maxy-y[0],miny-y[0],maxz-z[0],minz-z[0]]]
df=pd.DataFrame(data, columns = ['X','XX', 'Y','YY','Z','ZZ'])
df.insert(6, "key", str(pathex[0][-5]))
df
```

```python
for i in range(1,len(pathex)):
    dff=pd.read_csv(path+pathex[i])
    dff=dff.head(n)
    x=list(dff["X"])
    y=list(dff["Y"])
    z=list(dff["Z"])
    maxx=max(x)
    minx=min(x)
    maxy=max(y)
    miny=min(y)
    maxz=max(z)
    minz=min(z)
    data=[[maxx-x[0],minx-x[0],maxy-x[0],miny-x[0],maxz-x[0],minz-x[0]]]
    temp=pd.DataFrame(data, columns = ['X','XX', 'Y','YY','Z','ZZ'])
    temp.insert(6, "key", str(pathex[i][-5]))
    df=df.append(temp,ignore_index = True)
df
```

```python
x = df.iloc[:, 0:6].values
y = df.iloc[:, 6].values
```

```python
# x = df.drop('Home', axis=1)
# y = df['Home']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=1)
logmodel = LogisticRegression()
logmodel.fit(x_train, y_train)
 
predictions = logmodel.predict(x_test)
df = pd.DataFrame({'Actual': y_test, 'Predicted': predictions})
df
```
