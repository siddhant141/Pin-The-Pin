```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import csv
from sklearn.cluster import KMeans
```

```python
path='C:\\Users\\nilu1\\PatternCracking\\gyro\\9_Dec_2019_10_13_26.csv'
f=open(path)
para=f.readlines()
line=""
data=[]
for i in range(len(para)):
    line=para[i]
    line=line.split(",")
    for i in range(len(line)):
        line[i]=line[i].rstrip('\n')
        line[i]=line[i].strip('"')
        line[i]=float(line[i])
    line.pop(0)
    data.append(line)
X=np.array(data)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0],X[:,1],X[:,2], c='r', marker='o')
```

```python
kmeans = KMeans(n_clusters=10)
kmeans.fit(X)
#print(kmeans.cluster_centers_)
print(kmeans.labels_)
fig=plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0],X[:,1],X[:,2], c=kmeans.labels_, cmap='rainbow')
```

```python
fig=plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(kmeans.cluster_centers_[:,0] ,kmeans.cluster_centers_[:,1],kmeans.cluster_centers_[:,2], color='black')
```

```python
temp=kmeans.labels_
temp=np.ndarray.tolist(temp)
t=[]
for i in range(len(temp)):
    t.append([temp[i]])
```

```python
with open('label.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(t)
```
