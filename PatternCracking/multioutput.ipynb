```python
def setmultioutput(x):
    if(x==1):
        return [2,4,5]
    if(x==2):
        return [3,5,6]
    if(x==3):
        return [2,5,6]
    if(x==4):
        return [1,5,7]
    if(x==5):
        return [6,8,9]
    if(x==6):
        return [3,5,9]
    if(x==7):
        return [4,5,8]
    if(x==8):
        return [5,9,0]
    if(x==9):
        return [6,8,0]
    if(x==0):
        return [5,8,9]
```

```python
for file in pathex:
    ni=file.rfind("_")+1
    n=int(file[ni:].split(".")[0])
    df=pd.read_csv(path+file)
    x = df.iloc[:, 0:n].values
    y = df.iloc[:, n].values
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
    temp1=[]
    temp11=[]
    temp111=[]
    for i in range(len(y_train)):
        t=setmultioutput(y_train[i])
        temp1.append(t[0])
        temp11.append(t[1])
        temp111.append(t[2])
    temp2=[]
    temp22=[]
    temp222=[]
    for i in range(len(y_test)):
        t=setmultioutput(y_test[i])
        temp2.append(t[0])
        temp22.append(t[1])
        temp222.append(t[2])
    y_train=np.vstack((y_train, temp1, temp11,temp111)).T
    y_test=np.vstack((y_test, temp2, temp22,temp222)).T
#     model = RandomForestClassifier(n_estimators=75, random_state=0, max_depth=25)
#     model = KNeighborsClassifier(n_neighbors=1)
#     model = DecisionTreeClassifier(max_depth = 25)
    model=MLPClassifier(solver="lbfgs",max_iter=1500,warm_start=True,activation="identity",hidden_layer_sizes=(600,10))
    logmodel=[]
    logmodel = MultiOutputClassifier(model, n_jobs=-1)
    logmodel.fit(X_train, y_train)
#     y_pred = logmodel.predict(X_train)
#     print(file,end=" : ")
#     print(accuracy_score(y_train,y_pred))
    y_pred = logmodel.predict(X_test)
    count=0
    count1=0
    count2=0
    count3=0
    count4=0
    count12=0
    for i in range(len(y_pred)):
        if(y_test[i][0] in y_pred[i]):
            count+=1
        if(y_test[i][0] in y_pred[i][:2]):
            count12+=1
        if(y_test[i][0] == y_pred[i][0]):
            count1+=1
        elif(y_test[i][0] == y_pred[i][1]):
            count2+=1
        elif(y_test[i][0] == y_pred[i][2]):
            count3+=1
        elif(y_test[i][0] == y_pred[i][3]):
            count4+=1
#         print(y_test[i],y_pred[i])
    l=len(y_pred)
    print(l,count)
    print(count/l)
    print("At position 1 : ",count1/l)
    print("At position 2 : ",count2/l)
    print("At position 3 : ",count3/l)
    print("At position 4 : ",count4/l)
#     df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
#     score.append([f1_score(y_test, y_pred, average="micro"),file])
#     score.append([accuracy_score(y_test, y_pred),file])
    score.append([count1/l,file])
    score12.append([count12/l,file])
#     print(df)
score.sort(reverse=True)
score12.sort(reverse=True)
print("--------------------------------------------Test : -------------------------------------------")
for i in range(len(score)):
    print(score[i][1]+" : "+str(score[i][0]))
print("--------------------------------------------Test : -------------------------------------------")
for i in range(len(score)):
    print(score12[i][1]+" : "+str(score12[i][0]))
# df
```
