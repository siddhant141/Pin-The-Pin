```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from os import listdir
from os.path import isfile, join
```

```python
path="C:\\Users\\nilu1\\PatternCracking\\acc\\"
pathex=[]
temp = [f for f in listdir(path) if isfile(join(path, f))]
temp.sort()
for i in range(len(temp)):
    st=temp[i]
    if(st[:2]=="27"):
        pathex.append(temp[i])
# print(pathex)
```

```python
# n -> number of lines in csv file
# actual -> the actual key pressed
n=130
actual=1
x=[]
y=[]
z=[]
temp=pd.read_csv(path+pathex[0])
temp=temp.head(n)
x=list(temp["X"])
y=list(temp["Y"])
z=list(temp["Z"])
maxx=max(x)
minx=min(x)
maxy=max(y)
miny=min(y)
maxz=max(z)
minz=min(z)
data=[[maxx-x[0],minx-x[0],maxy-x[0],miny-x[0],maxz-x[0],minz-x[0]]]
df=pd.DataFrame(data, columns = ['X','XX', 'Y','YY','Z','ZZ'])
df.insert(6, "key", str(pathex[0][-5]))
df
```

```python
for i in range(1,len(pathex)):
    dff=pd.read_csv(path+pathex[i])
    dff=dff.head(n)
    x=list(dff["X"])
    y=list(dff["Y"])
    z=list(dff["Z"])
    maxx=max(x)
    minx=min(x)
    maxy=max(y)
    miny=min(y)
    maxz=max(z)
    minz=min(z)
    data=[[maxx-x[0],minx-x[0],maxy-x[0],miny-x[0],maxz-x[0],minz-x[0]]]
    temp=pd.DataFrame(data, columns = ['X','XX', 'Y','YY','Z','ZZ'])
    temp.insert(6, "key", str(pathex[i][-5]))
    df=df.append(temp,ignore_index = True)
df
```

```python
X = df.iloc[:, 0:6].values
y = df.iloc[:, 6].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
regressor = RandomForestRegressor(n_estimators=20, random_state=1)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
